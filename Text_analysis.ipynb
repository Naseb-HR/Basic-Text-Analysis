```
import pandas as pd
import numpy as np
import nltk
import os
import nltk.corpus
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```
f = open("Main_Draft.txt", "rb")
text = f.read().decode(errors='replace')
```

```
text
```

```
from nltk.tokenize import word_tokenize
tokenized_word=word_tokenize(text)
print(tokenized_word)
```

```
import re
```

```
tokens = re.findall('\w+', text)
tokens[:5]
```

```
# Import RegexpTokenizer from nltk.tokenize
from nltk.tokenize import RegexpTokenizer

# Create tokenizer
tokenizer = RegexpTokenizer('\w+')
```

```
tokens = tokenizer.tokenize(text)
tokens[:8]
```

```
from nltk.probability import FreqDist
fdist = FreqDist(tokens)
print(fdist)
```

```
import matplotlib.pyplot as plt
fdist.plot(30,cumulative=False)
plt.show()
```

```
# Initialize new list
words = []

# Loop through list tokens and make lower case
for word in tokens:
    words.append(word.lower())
```

```
words[:5]
```

```
sw = nltk.corpus.stopwords.words('english')
sw[:5]
```

```
# Initialize new list
words_ns = []

# Add to words_ns all words that are in words but not in sw
for word in words:
    if word not in sw:
        words_ns.append(word)
```

```
words_ns[:5]
```

```
from nltk.stem import PorterStemmer 
stemmer=PorterStemmer()

words_ns1 = []

for w in words_ns: 
  words_ns1.append(stemmer.stem(w)) 


```

```
#Import datavis libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Figures inline and set visualization style
%matplotlib inline
sns.set()

# Create freq dist and plot
freqdist1 = nltk.FreqDist(words_ns1)
freqdist1.plot(20)
```

```
from nltk.stem import WordNetLemmatizer 
# Create WordNetLemmatizer object 
wnl = WordNetLemmatizer() 

words_ns2 = []

for w in words_ns: 
  words_ns2.append(wnl.lemmatize(w)) 
```

```
#Import datavis libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Figures inline and set visualization style
%matplotlib inline
sns.set()

# Create freq dist and plot
freqdist1 = nltk.FreqDist(words_ns2)
freqdist1.plot(20)
```

```
from wordcloud import WordCloud, STOPWORDS 
```

```
 # Converts each token into lowercase 
comment_words = ''
for i in range(len(words_ns2)): 
        words_ns2[i] = words_ns2[i].lower() 
        
comment_words += " ".join(words_ns2)+" "
```

```

#wordcloud = WordCloud(background_color ='white').generate(str(words_ns2))

wordcloud = WordCloud(background_color ='white').generate(comment_words)

```

```
# plot the WordCloud image                        
plt.figure(figsize = (8, 9), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
